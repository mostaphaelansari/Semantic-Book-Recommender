**la distribution des notes**

```python
plt.figure(figsize=(8,5))
sns.histplot(books['average_rating'], bins=20, kde=True)
plt.xlabel("Note moyenne")
plt.ylabel("Nombre de livres")
plt.title("Distribution des notes moyennes des livres")
plt.show()
```

**Analyser la rÃ©partition des annÃ©es de publication**

```python
plt.figure(figsize=(8,5))
sns.histplot(books['published_year'].dropna(), bins=30, kde=True)
plt.xlabel("AnnÃ©e de publication")
plt.ylabel("Nombre de livres")
plt.title("RÃ©partition des annÃ©es de publication")
plt.show()
```

***Top 10 des livres les plus notÃ©s (en nombre dâ€™avis)***

```python
top_books = books.nlargest(10, 'ratings_count')

plt.figure(figsize=(10, 6))
sns.barplot(y=top_books['title'], x=top_books['ratings_count'], palette="coolwarm")
plt.xlabel("Nombre d'avis")
plt.ylabel("Titre du livre")
plt.title("Les 10 livres avec le plus de notes")
plt.show()
```

***Nombre de livres publiÃ©s par annÃ©e***


```python
plt.figure(figsize=(12, 5))
sns.histplot(books['published_year'].dropna(), bins=50, kde=True)
plt.xlabel("AnnÃ©e de publication")
plt.ylabel("Nombre de livres")
plt.title("RÃ©partition des annÃ©es de publication des livres")
plt.show()
```

***Relation entre le nombre de pages et la note moyenne***

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=books, x='num_pages', y='average_rating', alpha=0.5)
plt.xlabel("Nombre de pages")
plt.ylabel("Note moyenne")
plt.title("Relation entre le nombre de pages et la note moyenne")
plt.show()
```

 ***CatÃ©gories les plus frÃ©quentes***

```python
from collections import Counter

categories = books['categories'].dropna().str.split(';').explode()  # Si plusieurs catÃ©gories par livre, on les sÃ©pare
top_categories = pd.DataFrame(Counter(categories).most_common(10), columns=["CatÃ©gorie", "Nombre de livres"])

plt.figure(figsize=(10, 5))
sns.barplot(y=top_categories["CatÃ©gorie"], x=top_categories["Nombre de livres"], palette="magma")
plt.xlabel("Nombre de livres")
plt.ylabel("CatÃ©gorie")
plt.title("Top 10 des catÃ©gories les plus frÃ©quentes")
plt.show()
```

**Relation entre l'annÃ©e de publication et la note moyenne**

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=books, x='published_year', y='average_rating', ci=None)
plt.xlabel("AnnÃ©e de publication")
plt.ylabel("Note moyenne")
plt.title("Ã‰volution de la note moyenne des livres au fil du temps")
plt.show()
```

***CorrÃ©lations entre les variables numÃ©riques***

```python
plt.figure(figsize=(8, 6))
sns.heatmap(books[['average_rating', 'num_pages', 'ratings_count', 'published_year']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("CorrÃ©lation entre les variables numÃ©riques")
plt.show()
```
